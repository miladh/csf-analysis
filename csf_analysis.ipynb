```python
%matplotlib nbagg
from pylab import*
import pandas as pd
import seaborn
import glob
import scipy.optimize as cf



params = {'legend.fontsize': 20,
         'axes.labelsize': 20,
         'axes.titlesize': 20,
         'xtick.labelsize': 20,
         'ytick.labelsize':20,
         'lines.markersize': 9}
pylab.rcParams.update(params)
```

```python
data = {"1061": {}, 
        "1063": {}, 
        "1066": {}}


def dog_ft(k, a, b, w):
    return exp(-k**2 * a**2/4.) - w * exp(-k**2 * b**2/4.)

def dog(r, a, b, w):
    return exp(-r**2 / a**2)/pi/a**2 - w * exp(-r**2 / b**2)/pi/b**2


def bandpass_filter(k, a, b):
    return b - a * k**2

def lowpass_filter(k, a, b):
    return b * exp(a * k)
       
    
    
def fit_to_dog():
    print "Curve fitting to DoG function"
    for i, key in enumerate(data):
        for exp in data[key]:
            fit_params, fit_cov, rms_error, fitted_curve = fit_curve(frame=data[key][exp], curve=dog_ft)
            data[key][exp].dog_params = fit_params
            data[key][exp].dog_fit = fitted_curve
            data[key][exp].dog_rms = rms_error
            
            
def fit_to_bandpass():
    print "Curve fitting to bandpass filter"
    for i, key in enumerate(data):
        for exp in data[key]:
            fit_params, fit_cov, rms_error, fitted_curve = fit_curve(frame=data[key][exp], curve=bandpass_filter)
            data[key][exp].bandpass_params = fit_params
            data[key][exp].bandpass_fit = fitted_curve
            data[key][exp].bandpass_rms = rms_error

    

            
def fit_to_lowpass():
    print "Curve fitting to lowpass filter"
    for i, key in enumerate(data):
        for exp in data[key]:
            fit_params, fit_cov, rms_error, fitted_curve = fit_curve(frame=data[key][exp], curve=lowpass_filter)
            data[key][exp].lowpass_params = fit_params
            data[key][exp].lowpass_fit = fitted_curve
            data[key][exp].lowpass_rms = rms_error
            
        


def fit_curve(frame, curve):
    k = array(frame.condition)
    s = array(frame.sensivity)
    fit_params, fit_cov = cf.curve_fit(curve, k, s)
    fitted_curve = curve(k, *fit_params)
    rms_error =  compute_rms_error(fitted_curve, s)
    return fit_params, fit_cov, rms_error, fitted_curve 



def compute_rms_error(predictions, targets):
    return sqrt(mean((predictions-targets)**2))


def print_rms_errors():
    columns = ["DoG (C)", "Lowpass (C)", "Bandpass (C)", "DoG (A)", "Lowpass (A)", "Bandpass (A)" ]
    rms_frame = pd.DataFrame(columns=columns)
    for key in data:
        rms_errors = [NaN]*len(columns)
        for j, exp in enumerate(data[key]):
            rms_errors[j+2*j] = data[key][exp].dog_rms
            rms_errors[j+1+2*j] = data[key][exp].lowpass_rms
            rms_errors[j+2+2*j] = data[key][exp].bandpass_rms
            
        rms_frame.loc[key] = rms_errors
     
    
    mean_rms = zeros(len(columns))
    for i, col in enumerate(rms_frame.columns):
        mean_rms[i]=  mean(rms_frame[col])
    
    rms_frame.loc["Average rms"] = mean_rms
    display(rms_frame)
    


def read_csv_file(name):
    for csv_file in glob.glob(name+"*"):
        print csv_file
        id_name = csv_file.split("_")[1].split(".")[0]
        frame = pd.read_csv(csv_file, index_col=False, header=1)
        frame["sensivity"] = pd.Series(1./frame.mu)
        frame.id_name = id_name
        data[id_name][name] = frame
        
    return frame
    
```

```python
print "Reading csv files..."
read_csv_file("control")
read_csv_file("adaptation")


print "\nCurve fitting..."
fit_to_dog()
fit_to_bandpass()
fit_to_lowpass()


print "\nPrint root-mean-square error..."
print_rms_errors()
```
